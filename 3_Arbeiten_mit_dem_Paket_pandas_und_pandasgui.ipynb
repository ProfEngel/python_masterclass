## 3.2 Datensätze nach Spalten aufsplitten
Datensätze einlesen und im Anschluss in Features und Targets splitten? Das wird entweder per direktes Ansprechen der Spaltenbezeichnung oder per loc, iloc-Befehl gemacht.

```python
#Einen Datensatz aufsplitten in mehrere Features X und ein Target y
#siehe auch die Dokumentation: https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html

#Datensatz anzeigen lassen
df=pd.read_csv("../../Datasets/01.1_iris_full.csv", delimiter=",")
df.head()
```

```python
#In der Spalte species steht demnach das Target y
y= df[["species"]]
y.head()
#print(type(y))
```

```python
#In den Spalten nach Species stehen alle Features, welche unter X zusammengefasst werden
#Diese könnnen mittels iloc oder loc zusammengefasst an X übertragen werden.
#Siehe auch hier: https://www.delftstack.com/de/howto/python-pandas/select-multiple-columns-pandas/#:~:text=Um%20mehrere%20Spalten%20aus%20einem,iloc()%20und%20loc()%20.
#Soll per Index der Spalten alle Spalten nach der Spalte 0 in X geschrieben werden, dann wie folgt...
X= df.iloc[0:4,-1:]
X.head()
```

```python
#Sollen ebenfalls per Spaltenindex nur einzelne Spalten gewählt werden, dann so...
X1=df.iloc[:,[0,3]]
X1.head()
```

```python
#Sollen per Name der Spalten, entsprechende Spalten ausgewählt werden, dann wird der Befehl loc statt iloc verwendet
X2=df.loc[:,["sepal length", "sepal width"]]
X2.head()
```

## 3.3 Zeilen eines Dataframes ansprechen und List Slicing
Hin und wieder müssen einzelne Zeilen herausgegriffen werden. Dies kann ebenfalls mittels des iloc-Befehls gemacht werden.

```python
#gezielt nur den ersten Eintrag, aber vollständig ausgeben
#iloc steht für Index-Spalte ganz links
df_letzteZeile=df.iloc[149]
df_letzteZeile.head()
#print(df_letzteZeile)
```

```python
#Typ des Eintrags (Pandas-Klase) ermitteln
eintrag=df.iloc[0]
print(type(eintrag))
```

```python
#vorab den kompletten Eintrag ausgeben
print(df.iloc[2])
#gezielt auf einen Value eines Eintrags zugreifen
eintrag=df.iloc[2]
print("Der Eintrag für Petal Length lautet: ",eintrag["petal length"])
```

```python
#starten bei 4 und vor 8 beenden
df.iloc[4:8]
```

```python
#nur die letzten fünf Einträge ausgeben
df.iloc[-5:]
```

## 3.4 Pandas Daten verstehen

```python
# als Tupel ausgeben. Erster Wert ist die 0. der eigentliche Inhalt beginnt bei species bis zum Ende des Wertes in petal width
for row in df.iterrows():
    print(row)
    print(type(row))
    break
```

```python
# als Tupel ausgeben. Da es ein Tupel ist, kann gezielt der Tupelindex mit pos und der Inhalt mit d ausgegeben werden
for row in df.iterrows():
    pos=row[0]
    d=row[1]
    print(pos)
    print(d)
    break
```

```python
# als Tupel ausgeben. Da es ein Tupel ist, kann gezielt der Tupelindex mit pos und der Inhalt mit d ausgegeben werden
for pos,d in df.iterrows():
    #print(pos)
    print(d)
    break
```

```python
# gezielt die Species ausgeben
for pos,d in df.iterrows():
    print(d["species"])
```

## 3.7 Zeitreihen erstellen und abrufen
In Pandas können schnell Zeitreihen erzeugt und abgerufen werden. Dafür wird die Pandas-Funktion date_range() genutzt.

```python
#in dem folgenden Series-Objekt wird eine zweijähirige Zeitreihe eingefügt.
dates = pd.date_range('2021-01-01', '2022-12-31')

dates
```

```python
#Die Zeitschritte können ebenfalls definiert werden. Dafür benötigt man die Funktion freq. W = Wöchentlich, H = Stündlich
weekly = pd.date_range('2021-01-01', '2021-02-01', freq="W")

print(weekly)



hourly = pd.date_range('2021-01-01 8:00', '2021-01-01 18:00', freq="H")

print(hourly)
```
