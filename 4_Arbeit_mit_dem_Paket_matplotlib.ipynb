# 4. Matplotlib - Visualisierung von Daten in Python3
Um in Python3 Daten in Diagrammen zu visualisieren eignen sich mehrere Bibliotheken. Matplotlib, Seaborn sind die bekanntesten und auch die Bibliotheken, welche den größten Support bieten. Hier wird zunächst auf Matplotlib eingegangen.

Die genauen Befehle finden sich hier: https://matplotlib.org/stable/api/pyplot_summary.html

Einige davon werden hier angewendet.

Beispiele für was alles mit Matplotlib möglich ist: https://matplotlib.org/stable/gallery/index.html

```python
#Für ein erstes Beispiel werden zunächst die notwendigen Bibliotheken importiert
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# zwei Datenlisten erstellen
x = [1, 2, 1,4,5,10,3]
a = [2,3,4,5,6,7,8]
b = [1,4,3,4,2]
# Die Datenlisten gegenüberstellen (also den Wert x=2 dem Wert y=4 usw.)
plt.plot(x)
plt.plot(a)
plt.plot(b)
plt.title("Titel des Diagramms")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")
#plt.legend(["Das ist x/y"])
plt.show()
```

```python
# Jetzt noch ein bischen Datenbeschriftung hinzufügen
plt.plot(x, y)
plt.title("Unser Test")
plt.xlabel("Unser X-Wert")
plt.ylabel("Unser Y-Wert")
plt.legend(["Das ist x/y"])
plt.show()
```

```python
# und nun das Ganze als Säulendiagramm
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x,y)
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
# einfaches Bar-Chart
# Dies wird mittels dem Befehl ax.bar(x, height, width, bottom, align) umgesetzt.
# Siehe auch die Quelle: https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Vertiefungsmodule = ['VWL', 'IHL', 'DBM', 'Marketing', 'Personal']
Studenten = [23,17,35,29,12]
ax.bar(Vertiefungsmodule,Studenten)
plt.show()
```

```python
import matplotlib.pyplot as plt

data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

```python
#Die Bibliothek pandas importieren. Hier wird in der Regel folgender Befehl genutzt. Wir können damit unter pd die entsprechenden Pandas-Befehle nutzen
import pandas as pd
import matplotlib.pyplot as plt


#den bekannten IRIS-Datensatz aus 3. pandas einlesen
df=pd.read_csv("../../Datasets/01.1_iris_full.csv")
#nur die Spalte für das Target in y übertragen
y= df[["species"]]

#Verteilung der Inhalte im Target ausgeben
#df[["species"]].value_counts().plot(kind="bar");
y.value_counts().plot(kind="bar");

#Titel in der X-Achse leicht drehen
plt.xticks(rotation=25)
plt.show()
```

```python
# Vergleiche zwischen Daten je Abschnitt kann mittels gruppierten Säulen erreicht werden
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
```
